Que-2 How does broadcasting work in NumPy ?

Ans- In NumPy, broadcasting is a powerful feature that allows arithmetic operations on arrays of different shapes without the need for explicit reshaping or looping. It works by automatically expanding the smaller array so that it matches the shape of the larger array, enabling element-wise operations.

 How Broadcasting Works

When performing operations on two arrays, NumPy compares their shapes element-wise, starting from the trailing (rightmost) dimensions. Two dimensions are compatible when:

*  They are equal, or
*   of them is 1

If these conditions are not met, NumPy raises a ValueError indicating that the arrays could not be broadcast together.

**Broadcasting Rules**
1. Shape Alignment: If the arrays have different numbers of dimensions, the shape of the smaller array is padded with ones on the left side until both shapes have the same length.

2. Dimension Compatibility: Two dimensions are compatible if they are equal or if one of them is 1.

3. Resulting Shape: The resulting array shape is the maximum of the input shapes along each dimension.

7. Write a program to perform element-wise multiplication of two NumPy arrays .


```python
import numpy as np

def multiply_arrays(array1, array2):
    """
    Performs element-wise multiplication of two NumPy arrays.

    Args:
        array1 (np.ndarray): The first NumPy array.
        array2 (np.ndarray): The second NumPy array.

    Returns:
        np.ndarray: A new NumPy array containing the element-wise product of array1 and array2.
                      Returns None if the arrays have incompatible shapes.
    """
    if array1.shape != array2.shape:
        print("Error: Arrays must have the same shape for element-wise multiplication.")
        return None
    return np.multiply(array1, array2)

if __name__ == '__main__':
    array1 = np.array([[1, 2, 3], [4, 5, 6]])
    array2 = np.array([[7, 8, 9], [10, 11, 12]])

    result_array = multiply_arrays(array1, array2)

    if result_array is not None:
        print("Array 1:")
        print(array1)
        print("\nArray 2:")
        print(array2)
        print("\nElement-wise multiplication result:")
        print(result_array)
```

1. Using matmul():
numpy.matmul(a, b): This function performs matrix multiplication of two arrays, a and b.


```python
# Example
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = np.matmul(a, b)
print(c)
```

2. Using the @ operator:
a @ b: This is a shorthand for np.matmul(a, b) and performs matrix multiplication.

```python
# Example
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = a @ b
print(c)
```

3. Element-wise multiplication using multiply():

 numpy.multiply(a, b): This function performs element-wise multiplication of two arrays, a and b.

```python
# Example
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = np.multiply(a, b)
print(c)
```

4. Element-wise multiplication using the * operator:

 a * b: This operator performs element-wise multiplication if both a and b are arrays.

```python
# Example
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = a * b
print(c)
```

12.  Create a 3D scatter plot using Plotly.

To create a 3D scatter plot using Plotly in Python, you can utilize either the high-level plotly.express interface or the more customizable plotly.graph_objects. Below are examples of both approaches:

Option 1: Using plotly.express (Simplified Interface)


```python
import plotly.express as px

# Load a sample dataset
df = px.data.iris()

# Create a 3D scatter plot
fig = px.scatter_3d(
    df,
    x='sepal_length',
    y='sepal_width',
    z='petal_width',
    color='species',  # Color by species
    symbol='species',  # Different symbols for each species
    size='petal_length',  # Size of markers based on petal length
    opacity=0.7
)

# Display the plot
fig.show()
```

Option 2: Using plotly.graph_objects (More Customization)

```python
import plotly.graph_objects as go
import numpy as np

# Generate sample data
t = np.linspace(0, 10, 100)
x = np.cos(t)
y = np.sin(t)
z = t

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=z,  # Set color to the z-values
        colorscale='Viridis',  # Choose a colorscale
        opacity=0.8
    )
)])

# Display the plot
fig.show()
```
