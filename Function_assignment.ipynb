Que-1 What is the difference between a function and a method in Python ?

Ans- In Python, both functions and methods are blocks of reusable code, but they differ primarily in their association with classes and objects. Here's a concise comparison:â€‹

** Function**

**Definition :** A function is a block of code defined using the def keyword, which can be called independently.

**Scope :** Functions are not tied to any class or object.

**Invocation :** Called by their name directly.

**Parameters :** Do not require self as the first parameter.

Example:â€‹

```python
def add(x, y):
    return x + y

result = add(5, 3)
print(result)
```

**Method**

**Definition :** A method is a function defined within a class and is associated with an object of that class.

**Scope :** Methods are bound to the class and can access and modify the object's attributes.

**Invocation :** Called on an object using dot notation.

**Parameters :** The first parameter is always self, which refers to the instance of the class.

Example:â€‹

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print("Woof!")

my_dog = Dog("Buddy")
my_dog.bark()
print(my_dog.name)
```

Que-2  Explain the concept of function arguments and parameters in Python.

Ans- In Python, functions are blocks of reusable code that can accept inputs to perform specific tasks. These inputs are known as parameters and arguments, and understanding their roles is crucial for effective programming.

**Parameters vs. Arguments**

Parameters are variables defined in a function's declaration. They act as placeholders for the values that will be provided when the function is called. For example:

```python
def greet(name):   # 'name' is a parameter
  print("Hello, " + name + "!")
```

Arguments are the actual values passed to the function when it is called. They
replace the parameters during execution. For instance:


```python
greet("Prince")    # "Prince" is an argument
```

In summary, parameters are part of the function's definition, while arguments are provided during the function call .

ðŸ”¹ **Types of Arguments in Python**

Python offers several ways to pass arguments to functions, enhancing flexibility and readability:



**1.Positional Arguments :** The order of arguments matters. Each argument is assigned to the corresponding parameter based on its position.



```python
def describe_pet(name, species):
    print(f"{name} is a {species}")

describe_pet("Buddy", "dog")
```

**2.Keyword Arguments :** Arguments are passed by explicitly naming the parameter, allowing them to be provided in any order.


```python
def introduce(name, age, city):
    print(f"My name is {name}, I am {age} years old, and I live in {city}.")

introduce(age=25, city="New York", name="Alice")
```

**3. Default Arguments :** Parameters can have default values. If an argument is not provided during the function call, the default value is used.


```python
def greet(name="there"):
    print(f"Hello, {name}!")

greet()           # Uses default value
greet("Alice")    # Overrides default value
```

**4. Variable-Length Arguments :** Sometimes, you may not know in advance how many arguments will be passed. Python provides *args and **kwargs for such cases:

*args: Allows passing a variable number of positional arguments.

**kwargs: Allows passing a variable number of keyword arguments.


```python
def sum_numbers(*args):
    return sum(args)

print(sum_numbers(1, 2, 3, 4))
```

Que-6. Explain the concept of generators in Python and how they are defined ?

Ans- Generators are a special type of iterator that allow you to generate values on the fly without storing them all in memory. They're useful for working with large datasets or streams of data.


**Key Characteristics:**

*   Defined using functions with the yield keyword.
*   Maintain state between iterations.
*   Automatically create an iterator object.
*   More memory-efficient than lists for large data.

Defining a Generator

Use yield instead of return:



```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
```

Using the Generator

```python
counter = count_up_to(3)

print(next(counter))  # 1
print(next(counter))  # 2
print(next(counter))  # 3
# next(counter) -> raises StopIteration
```

Que-7. What are the advantages of using generators over regular functions ?

Ans- **Advantages of Using Generators Over Regular Functions**

1. Memory Efficiency

   Generators yield items one at a time, rather than creating and storing the entire result in memory.

 ðŸ”¹ Useful for processing large datasets or infinite sequences.

2. Lazy Evaluation

  Values are computed only when needed, which can significantly reduce computation time in certain cases.

3. Simplified Code

 Generators can replace complex iterator classes with concise, readable syntax using yield.

4. Maintains State Automatically

 No need to manually track iteration state (like counters or indexes); generators pause and resume execution at each yield.

5. Pipelining Capabilities

 Generators can be chained together to form efficient processing pipelines (like Unix pipes), reducing overhead.

Que-9. Explain the purpose and usage of the `map()` function in Python.

Ans- The map() function in Python serves to apply a given function to each item of an iterable (e.g., list, tuple) and returns an iterator that yields the results. It provides a way to perform operations on all elements of an iterable without explicit loops.

```python
# For example, to square each number in a list:

numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))
```

Que- 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?

Ans- The map(), filter(), and reduce() functions are built-in higher-order functions in Python that operate on iterables. They each serve different purposes:

map():

Applies a given function to each item in an iterable (e.g., list, tuple) and returns an iterator that yields the results. It transforms each element individually.

filter():

Creates a new iterable containing only the items from the original iterable for which a given function returns True. It selectively keeps elements based on a condition.

reduce():

Applies a function cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It aggregates elements. It requires importing from the functools module in Python 3.

```python
numbers = [1, 2, 3, 4, 5]

# map(): double each number

squared_numbers = list(map(lambda x: x * 2, numbers))
print(squared_numbers)
```

```python
# filter(): keep only even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
 from functools import reduce  # Import reduce first

numbers = [1, 2, 3, 4, 5]

# Now reduce works correctly
total = reduce(lambda x, y: x + y, numbers)

print("Sum:", total)  # Output: Sum: 15
```

```python
 def sum_of_even_numbers(numbers):
    even_sum = 0  # Initialize the sum to 0
    for num in numbers:
        if num % 2 == 0:  # Check if the number is even
            even_sum += num  # Add the even number to the sum
    return even_sum
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2 or n > 200:
        return False  # Not prime if less than 2 or greater than 200
    for i in range(2, int(n ** 0.5) + 1):  # Only check up to square root of n
        if n % i == 0:
            return False  # Found a factor, so not prime
    return True  # No factors found, so it's prime
```

```python
print(is_prime(7))     # True
print(is_prime(100))   # False
print(is_prime(199))   # True
print(is_prime(201))   # False (out of range)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total number of terms to generate
        self.count = 0          # Counter to keep track of generated terms
        self.a = 0              # First term in the Fibonacci sequence
        self.b = 1              # Second term in the Fibonacci sequence

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration  # No more terms to generate
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
```

```python
# Example usage:
n = 10  # Specify the number of terms
fib_iterator = FibonacciIterator(n)
for num in fib_iterator:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    """
    A generator function that yields powers of 2 up to the given exponent.

    :param exponent: The maximum exponent to yield powers of 2.
    """
    for i in range(exponent + 1):
        yield 2 ** i  # Yield 2 raised to the power of i
```

```python
for power in powers_of_two(5):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string

> Add blockquote



```python
def read_file_by_line(filepath):
    """
    Reads a file line by line and yields each line.

    Args:
        filepath (str): The path to the file.

    Yields:
        str: Each line of the file.
    """
    with open(filepath, 'r') as file:
        for line in file:
            yield line
```

```python
filepath = 'my_file.txt'
with open(filepath, 'w') as f:
    f.write("This is the first line.\n")
    f.write("This is the second line.\n")
    f.write("This is the third line.\n")

for line in read_file_by_line(filepath):
    print(line, end='')
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    # Use filter to keep only non-vowel characters
    return ''.join(filter(lambda char: char not in vowels, input_str))

# Example usage
text = "Hello, World!"
result = remove_vowels(text)
print("String without vowels:", result)
```
